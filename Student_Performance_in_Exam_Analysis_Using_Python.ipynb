#Student Performance in Exam Analysis Using Python

# Project Description:


This project analyzes student performance in math, reading, and writing based on demographic and socio-economic factors such as gender, race/ethnicity, parental education level, lunch type, and test preparation course participation. By examining these attributes, we aim to identify patterns and insights that can help educators and policymakers improve student outcomes and develop more effective educational strategies.

# Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load dataset

```python
df = pd.read_csv('14_Student Performance in Exam Analysis.csv')
```

# Perform basic data exploration

```python
print(df.head())
print(df.info())
print(df.describe())
```

# Load Columns

```python
print(df.columns)
```

# Descriptive Statistics

```python
print("\nDescriptive Statistics:")
print(df.describe(include='all'))
```

# Visual Insights:

 # Plot histograms of math, reading, and writing scores

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.hist(df['math score'], bins=10, color='blue', alpha=0.7)
plt.title('Math Score Distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')
```

```python
plt.subplot(1, 3, 2)
plt.hist(df['reading score'], bins=10, color='green', alpha=0.7)
plt.title('Reading Score Distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')
```

```python
plt.subplot(1, 3, 3)
plt.hist(df['writing score'], bins=10, color='red', alpha=0.7)
plt.title('Writing Score Distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')
```

```python
plt.tight_layout()
plt.show()
```

# Bar plots for gender distribution in scores

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='gender', y='math score', data=df, ci=None)
plt.title('Average Math Score by Gender')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='gender', y='reading score', data=df, ci=None)
plt.title('Average Reading Score by Gender')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='gender', y='writing score', data=df, ci=None)
plt.title('Average Writing Score by Gender')
plt.show()
```

# Impact of parental education

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='parental level of education', y='math score', data=df, ci=None)
plt.xticks(rotation=45)
plt.title('Average Math Score by Parental Education Level')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='parental level of education', y='reading score', data=df, ci=None)
plt.xticks(rotation=45)
plt.title('Average Reading Score by Parental Education Level')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='parental level of education', y='writing score', data=df, ci=None)
plt.xticks(rotation=45)
plt.title('Average Writing Score by Parental Education Level')
plt.show()
```

Summary of Key Findings:
1. Female students generally perform better in reading and writing compared to male students.
2. Higher parental education levels are associated with higher student performance in all subjects.
3. Students who receive standard lunch tend to score higher than those who receive free/reduced lunch.
4. Completing the test preparation course appears to improve performance in reading and writing.
5. There are notable differences in performance among different racial/ethnic groups, which may indicate underlying socio-economic or educational disparities.


Conclusion:

The analysis provides valuable insights into factors affecting student performance. These insights can help educators and policymakers design targeted interventions to improve student outcomes. Emphasizing data-driven decision-making can lead to more effective educational strategies.
